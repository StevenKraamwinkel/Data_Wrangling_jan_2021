## Extracting energy use per capita and CO2 emissions from the USA 

```python
from_usa_data = global_data_101.loc[global_data_101['Country name']=='United States'].reset_index()
```

```python
a = from_usa_data.iloc[[6,7],[5,8,13,18,23,28]]
a.iloc[1,5] = 'NaN' 
```

## Creating dataframe from the extracted data

```python
dataframe = pd.DataFrame({'Series name': ['Energy use per capita (tonne of oil equivalent)', 'CO2 emissions, (KtCO2)'],
                        '1990': [a.iloc[0,1], a.iloc[1,1]],
                          '1995': [a.iloc[0,2], a.iloc[1,2]],
                          '2000': [a.iloc[0,3], a.iloc[1,3]],
                          '2005': [a.iloc[0,4], a.iloc[1,4]],
                          '2010': [a.iloc[0,5], a.iloc[1,5]],
                        })
dataframe.dtypes
```

```python
#setting all column data from string to float datatype
df = pd.DataFrame(dataframe)
df['1990'] = pd.to_numeric(df['1990'], errors='coerce')
df['1995'] = pd.to_numeric(df['1995'], errors='coerce')
df['2000'] = pd.to_numeric(df['2000'], errors='coerce')
df['2005'] = pd.to_numeric(df['2005'], errors='coerce')
df['2010'] = pd.to_numeric(df['2010'], errors='coerce')

df.dtypes
```

```python
pd.options.display.float_format = "{:,.2f}".format
```

### Setting first column as index and dataframe display 

```python
new_df = dataframe.set_index('Series name')
display(new_df)
```

### Int conversion of dataframe columns

```python
for i in new_df.columns:
    try:
        new_df[[i]] = new_df[[i]].astype(float).astype(int)
    except:
        pass
print(new_df.dtypes)
```

```python
new_df.iloc[0,4] = int(new_df.iloc[0,4])
display(new_df)
```
